```python
df = pd.read_csv('Airplane_Crashes_and_Fatalities.csv')
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.drop('Flight #',axis=1)
```

```python
from sklearn.impute import SimpleImputer
se = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
se.fit(df[['AC Type']])
df[['AC Type']] = se.transform(df[['AC Type']])
```

```python
df.isnull().sum()
```
